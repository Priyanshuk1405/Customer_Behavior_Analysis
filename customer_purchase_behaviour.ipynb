```python
def fillna_with_median(x):
    return x.fillna(x.median()) 

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(fillna_with_median)    
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# For Easy Reading changing column names into lower case and repalcing spaces with _ .
df.columns = df.columns.str.lower()
df.columns
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# Creating a column of age_group
label = ['Young', 'Adult', 'Middle-age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = label)
```

```python
df[['age','age_group']].tail(10)
```

```python
# Purchase Frequency Days
frequency_map = {
    'Fortnightly':14,
    'Weekly': 7,
    'Quarterly':90, 
    'Annually':365,
    'Bi-Weekly':14,
    'Every 3 Months':90,
    'Monthly':30
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_map)
df.head()
```

```python
# Now Comparing Both promo code and discount applied fro checking redundancy 
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop(columns='promo_code_used', axis=1)
```
